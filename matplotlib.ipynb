# Lets Explore Matplotlib 

```python
from matplotlib import pylab
print(pylab.__version__)
```

```python
# use numpy to generate randam data
import numpy as np
x=np.linspace(0,10,25)
y=x*x+2
print(x)
print(y)
```

# Operator Description

```python
from matplotlib import pyplot as plt
```

```python
fig=plt.figure()
axes=fig.add_axes([0.5,0.1,0.8,0.8,])
axes.plot(x,y,'r')
```

```python
# again we can draw subgraphs
fig,axes=plt.subplots(nrows=1,ncols=2)
for ax in axes:
    ax.plot(x,y,'r')
```

```python
# we can also draw a pitcure, or graph,inside another graph
fig=plt.figure(figsize=(6,6))
axes1=fig.add_axes([0.1,0.1,0.8,0.8])# big axes
axes2=fig.add_axes([0.3,.4,.4,0.3])# small canvas
axes1.plot(x,y,'r')
axes2.plot(y,x,'g')
```

```python
fig,axes=plt.subplots(dpi=200)
axes.plot(x,y,'p')
axes.set_xlabel('X Axis')
axes.set_ylabel("Y Axis")
axes.set_title("title of graph")
axes.plot(x,x**2,'r') 
axes.plot(x,x**3,'c') 

axes.legend(["Y= X**2","Y=X**3"],loc=2)
```

```python
fig, axes=plt.subplots(dpi=150)
axes.plot(x,x+1,'r')
axes.plot(x,x+2)
axes.plot(x,x+3)
axes.plot(x,x+4)

```

```python
fig,axes=plt.subplots(dpi=150)
axes.plot(x,x+1,color='blue',linewidth=0.25)
axes.plot(x,x+2,color='blue',linewidth=1)
axes.plot(x,x+3,color='blue',linewidth=1.5)
axes.plot(x,x+4,color='blue',linewidth=2)
```

# Other 2D graphics

```python
n=np.array([0,1,2,3,4,5])
fig,axes=plt.subplots(1,4,figsize=(16,5))
axes[0].set_title("Scatter")
axes[0].scatter(x,x+0.25*np.random.randn(len(x)))

axes[1].set_title("step")
axes[1].step(n,n**2,lw=3)

axes[2].set_title("Bar")
axes[2].bar(n,n**2,align='center',alpha=0.5)

axes[3].set_title("Fill Between")
axes[3].fill_between(x,x**2,color='purple',alpha=0.5)
```

```python
# Draw a redar chart
fig=plt.figure(figsize=(6,6))
ax=fig.add_axes([0.0,0.0,.6,.6],polar=True)
t=np.linspace(0,2*np.pi,100)
ax.plot(t,t*5,color='orange',lw=7)
```

```python
# Draw a histogram
n=np.random.randn(100000)
fig,axes=plt.subplots(1,2,figsize=(12,4))
axes[0].set_title("Default Histogram")
axes[0].hist(n)

axes[1].set_title("Cumulative Deteiled Histogram")
axes[1].hist(n,cumulative=True,bins=30)
```

```python
# Draw Contour image
import matplotlib.cm as cm
delta =0.025
x=np.arange(-3.0,3.0,delta)
y=np.arange(-2.0,2.0,delta)
X,Y=np.meshgrid(x,y)
Z1=np.exp(-X**2-Y**2)
Z2=np.exp(-(X-1)**2-(Y-1)**2)
Z=(Z1-Z2)*2
print(X)
print(Y)
```

```python
fig,ax=plt.subplots()
CS=ax.contour(X,Y,Z)
ax.clabel(CS,inline=1,fontsize=10)
ax.set_title('contour map')
```

```python
# Draw a #D image
from mpl_toolkits.mplot3d.axes3d import Axes3D

fig =plt.figure(figsize=(14,6))
#Specify the #D graphics to with projection '3d'
ax=fig.add_subplot(1,2,1,projection='3d')
ax.plot_surface(X,Y,Z,rstride=4,cstride=4,linewidth=0)
```

# write a python program to create a pie chart of the popularity of programming languages



```python
import matplotlib.pyplot as plt
# Data to plot
languages='Python','Java',"PHP",'Javascript','C#','C++'
popularity=[22.2,17.6,8.8,8,7.7,6.7,]
colors=["#1f77b4","#ff7f0e","#2ca02c","#d62728","#9467bd","#8c564b"]
```

```python
#explode 1st slic
explode=(0.1,0,0,0,0,0)
plt.pie(popularity,explode=explode,labels=languages,colors=colors,
autopct="%1.1f%%",shadow=True,startangle=140)
```
