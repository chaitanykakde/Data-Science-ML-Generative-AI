# Series Operations

```python
arr1=[0,1,2,3,4,5,7]
arr2=[6,7,8,9,5]
```

```python
s5=pd.Series(arr2)
s5
```

```python
s6=pd.Series(arr1)
s6
```

```python
s5.add(s6)
```

```python
s5.sub(s6)
```

```python
s7=s5.mul(s6)
s7
```

```python
s5.div(s6)
```

```python
print("Median:",s7.median())
print("Max:",s7.max())
print("Min:",s7.min())
print("Mode:",s7.mode())
print("Mean:",s7.mean())
```
